## 라이브러리 불러오기

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import pickle
```

## Inference용 데이터 불러오기

```python
test = pd.read_csv("/data/공모전데이터셋/round1_test.csv")
```

```python
test
```

## 학습한 모델 가져오기 (feature 생성 모델)

```python
# pickle을 사용해 모델 불러오기

from joblib import dump, load
load_model_lr = load('/result/model_lr.joblib') 

new_feat = load_model_lr.predict(test[['P1','P2','P3','P4']])
new_feat
```

```python
# 새로운 feature 1 행 추가
test['new_feature'] = new_feat
```

```python
# 새로운 feature 2 행 추가
test['new_feature2'] = 0
test.loc[(test['P2'] > test['P4']) | (test['P3'] > test['P4']), 'new_feature2'] = 0
test.loc[(test['P2'] < test['P4']) & (test['P3'] < test['P4']), 'new_feature2'] = 1
```

```python
# 새로운 feature 3, 4 행 추가
test['P4_P3'] = test['P4'] - test['P3']
test['P4_P2'] = test['P4'] - test['P2']
```

```python
test
```

## 데이터 전처리

```python
# ID관련 column, 시각, P5, 광고 응답 소재 카테고리 행 삭제
test.drop(columns=['시각', 'ADID', 'DSP ID', '매체 ID', '애드유닛 ID', '노출 ID', 'SSP 입찰ID', 'DSP 입찰ID', 'AX 낙찰ID', 'WUID (웹 유저 ID)', 'OS 버전 ID', '국가코드 ID', '광고 응답 소재 카테고리'], inplace=True)
```

```python
# 결측치 확인
test.isna().sum()
```

```python
# 광고 응답 광고주 도메인이 너무 많은 범주를 가지므로 제거
test.drop(columns=['광고 응답 광고주 도메인'], inplace=True)

test
```

```python
# 결측치 확인
test.isna().sum()
```

```python
# 가변수화
test_d = pd.get_dummies(test, columns=['ADID 타입', '플랫폼', 'OS 종류', '사이즈 ID'], drop_first=True)
```

```python
# 행 이름 확인
list(test_d)
```

```python
test_d.shape
```

```python
# model에 넣을 변수만 추출
test_x = test_d.drop(columns='Bid Index')
```

## 학습한 모델 불러오기 & 예측값 추출

```python
from joblib import dump, load
load_model_rf = load('/result/model_rf.joblib') 
pred_rf = load_model_rf.predict(test_x)
```

```python
test_d['pred'] = pred_rf
test_d
```

```python
# 예측값과 Bid index만 추출
result = test_d[['Bid Index', 'pred']]
result
```

```python
result.to_csv("과제2_예선_예측결과_JS.csv", index=False)
```
