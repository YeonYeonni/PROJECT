```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 등산로_엑셀

```python
df = pd.read_csv("./data/등산로_엑셀.csv", encoding="cp949")
df = df.set_index('산코드')
df = df[['산명', '등산로길이(km)', '등산로난이도명', '등산로상행시간', '등산로하행시간']]
```

```python
df_re = df.groupby(['산명'], as_index=False).mean()
df_re = df_re.set_index('산명')
```

```python
df_re['등산로난이도명'] = df_re['등산로난이도명'].round(0).astype(str)
df_re.info()
```

```python
df_re.tail(325)
```

```python
df_fn = df_re.loc[['가야산', '계룡산', '덕유산_향적봉', '무등산', '북한산', '설악산_대청봉', '소백산_비로봉', '속리산', '오대산', '오대산_노인봉', '월악산_영봉', '주왕산', '지리산_천왕봉'], :]

df_fn = df_fn.rename(index = {'덕유산_향적봉' : '덕유산', '설악산_대청봉' : '설악산', '소백산_비로봉' : '소백산', '오대산_노인봉' : '오대산', '월악산_영봉' : '월악산', '지리산_천왕봉' : '지리산'})
df_fn
```

# 국립공원 등산로별 평균 칼로리 소모량

```python
df2 = pd.read_csv("./data/국립공원/국립공원 등산로별 평균 칼로리 소모량.csv", encoding="cp949")
```

```python
df2 = df2[['trk_idno', 'ntnl_park_nm', 'trk_lngth', 'trk_rqrmn_tm', 'sxdst_cd', 'aggrp_sctin_val', 'ntnl_park_sctin_nm', 'trk_avrg_kcal_cnsmp_qntt']] # column 추출
df2 = df2.rename(columns = {'trk_idno':'고유번호'}) # column 이름 변경
df2 = df2.rename(columns = {'ntnl_park_nm':'산명'})
df2 = df2.rename(columns = {'trk_lngth':'트랙길이'})
df2 = df2.rename(columns = {'trk_rqrmn_tm':'트랙소요시간'})
df2 = df2.rename(columns = {'sxdst_cd':'성별'})
df2 = df2.rename(columns = {'aggrp_sctin_val':'나이대'})
df2 = df2.rename(columns = {'ntnl_park_sctin_nm':'국립공원구분명'})
df2 = df2.rename(columns = {'trk_avrg_kcal_cnsmp_qntt':'평균칼로리소모량'})
df2 = df2.set_index('고유번호') # index '고유번호'로 변경
```

```python
df2 = df2.groupby(['산명'], as_index=False).mean()
df2 = df2.set_index('산명')
```

```python
df2 = df2.drop('경주',axis=0)
df2 = df2.drop('내장산',axis=0)
df2
```

# 데이터프레임 병합

```python
df3 = pd.concat([df_fn, df2], axis = 1)
df3 = df3.rename(columns = {'등산로길이(km)':'길이'})
df3 = df3.rename(columns = {'등산로난이도명':'난이도'})
df3 = df3.rename(columns = {'등산로하행시간':'하행시간'})
df3 = df3.rename(columns = {'등산로상행시간':'상행시간'})

df3
```

```python
df3.to_csv('sample.csv',mode="w",encoding='euc-kr')
```
