```python
from PIL import Image, ImageEnhance, ImageChops
import numpy as np
import random

##########데이터 로드
mylist = str(0) #늘리고자 하는 이미지 데이터 숫자

image = Image.open("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+".jpg") #이미지 데이터 파일 위치
image = image.convert('RGB') #'L': greyscale, '1': 이진화, 'RGB' , 'RGBA', 'CMYK'         
image = image.resize((300, 300))

#밝기
enhancer = ImageEnhance.Brightness(image)
brightness_image = enhancer.enhance(1.8)
brightness_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_brightness.jpg")
                      
#좌우 대칭
horizonal_flip_image = image.transpose(Image.FLIP_LEFT_RIGHT) 
horizonal_flip_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_horizonal_flip.jpg")
                          
#상하 대칭
vertical_flip_image = image.transpose(Image.FLIP_TOP_BOTTOM) 
vertical_flip_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_vertical_flip.jpg")
                         
#좌우 이동
width, height = image.size
shift = random.randint(0, width * 0.2)
horizonal_shift_image = ImageChops.offset(image, shift, 0)
horizonal_shift_image.paste((0), (0, 0, shift, height))
horizonal_shift_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_horizonal_shift.jpg")
                           
#상하 이동
width, height = image.size
shift = random.randint(0, height * 0.2)
vertical_shift_image = ImageChops.offset(image, 0, shift)
vertical_shift_image.paste((0), (0, 0, width, shift))
vertical_shift_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_vertical_shift.jpg")
                          
#회전 
rotate_image = image.rotate(random.randint(-30, 30))
rotate_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_rotate.jpg")
                  
#기울기
#cx, cy = 0.1, 0
#cx, cy = 0, 0.1
cx, cy = 0, random.uniform(0.0, 0.3)
shear_image = image.transform(
    image.size,
    method=Image.AFFINE,
    data=[1, cx, 0,
          cy, 1, 0,])
shear_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_shear.jpg")
                 
#확대 축소
zoom = random.uniform(0.7, 1.3) #0.7 ~ 1.3
width, height = image.size
x = width / 2
y = height / 2
crop_image = image.crop((x - (width / 2 / zoom), y - (height / 2 / zoom), x + (width / 2 / zoom), y + (height / 2 / zoom)))
zoom_image = crop_image.resize((width, height), Image.LANCZOS)
zoom_image.save("C:\\Users\\hyk\\Desktop\\p_and_n_4\\p_and_n_4\\train\\n\\은행나무."+mylist+"_zoom.jpg")
```
