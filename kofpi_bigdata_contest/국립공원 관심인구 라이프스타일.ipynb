```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#국립공원 정보획득자(검색/블로그/뉴스기사등)의 DEMO 정보

df1 = pd.read_csv("./이정연 데이터 분석/data/국립공원/국립공원 관심인구 라이프스타일 (온라인쇼핑몰 이용 카테고리 정보).csv", encoding="cp949")
df1 = df1.drop(['frnt_ntds_prfn_cont', 'dgtl_hmpl_prfn_cont', 'lsur_car_prfn_cont', 'bty_prfn_cont', 'lilhd_kchn_prfn_cont', 'sprt_fshn_prfn_cont', 'grcs_helth_prfn_cont',
                'jrny_prfn_cont', 'fshn_wear_prfn_cont', 'auser_fshn_stff_prfn_cont', 'info_obtn_site_url'], axis=1)
df1 = df1.drop_duplicates(keep='first') #똑같은 행은 첫번째를 남기고 제거 -> 중복 제거
df1
```

```python
df1 = df1.drop(['auser_ara_gu_nm', 'auser_ara_dong_nm', 'car_group_nm', 'craln_tlphn_mkr_nm', 'craln_tlphn_sbscr_cmpn_nm'], axis=1)
df1
```

```python
#성별 분류
cond_gender = []
cond_gender.append(df1['auser_sxdst_cd'] == 'F')
cond_gender.append(df1['auser_sxdst_cd'] == 'M')
df1_f = df1[cond_gender[0]]
df1_m = df1[cond_gender[1]]
```

```python
#나이 분류
cond_age = []
for i in range(2, 7):
    cond_age.append(df1['age_group_cd'] == i)
```

```python
#직업

occpn_list = list(df1['occpn_nm'].unique())
occpn_list

for occpn in occpn_list:
    temp_list = []
    temp_cond = df1['occpn_nm'] == occpn
    temp_df = df1[temp_cond]
    temp_list = temp_df['lfst_cd'].value_counts()
    
    print("<", occpn, ">")
    print(temp_list)
    
    #자체적인 서버에서 진행해 글자가 깨집니다.
    #temp_list.plot(kind = 'pie', autopct = lambda p : '{:.2f}%'.format(p))
    #plt.legend()
    #plt.show()
    
    print()
```
