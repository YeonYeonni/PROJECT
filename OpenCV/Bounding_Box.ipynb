```python
import json
import cv2
import os

json_path = './json/R_020_50_M'
img_path = './raw_data/R_020_50_M'

json_list = os.listdir(json_path)
img_list = os.listdir(img_path)

for i in range(len(json_list)):
    
    #print(json_list[i])
    #print(img_list[i])
    
    json_direct = json_path + '/' + json_list[i]
    img_direct = img_path + '/' + img_list[i]
    
    #print(json_direct)
    #print(img_direct)
    
    json_name = '{}'.format(json_direct)
    img_name = '{}'.format(img_direct)
    
    #print(json_name)
    #print(img_name)
    
    # json 읽기
    with open(json_name) as data_file:
        local = json.load(data_file)
        
    # 이미지 읽기
    sample = cv2.imread(img_name)
        
    # json에서 좌표 검출
    Face = local['ObjectInfo']['BoundingBox']['Face']['Position']
    Face = list(map(float, Face))
    Face = list(map(int, Face))

    Leye = local['ObjectInfo']['BoundingBox']['Leye']['Position']
    Leye = list(map(float, Leye))
    Leye = list(map(int, Leye))

    Reye = local['ObjectInfo']['BoundingBox']['Reye']['Position']
    Reye = list(map(float, Reye))
    Reye = list(map(int, Reye))

    Mouth = local['ObjectInfo']['BoundingBox']['Mouth']['Position']
    Mouth = list(map(float, Mouth))
    Mouth = list(map(int, Mouth))

    Cigar = local['ObjectInfo']['BoundingBox']['Cigar']['Position']
    Cigar = list(map(float, Cigar))
    Cigar = list(map(int, Cigar))

    Phone = local['ObjectInfo']['BoundingBox']['Phone']['Position']
    Phone = list(map(float, Phone))
    Phone = list(map(int, Phone))
    
    # 사각형 그리기
    if Face[0] + Face[1] != 0:
        cv2.rectangle(sample, (Face[0], Face[1]), (Face[2], Face[3]), (255, 255, 255), 3)
    
    if Leye[0] + Leye[1] != 0:
        cv2.rectangle(sample, (Leye[0], Leye[1]), (Leye[2], Leye[3]), (0, 0, 255), 3)

    if Reye[0] + Reye[1] != 0:
        cv2.rectangle(sample, (Reye[0], Reye[1]), (Reye[2], Reye[3]), (0, 255, 0), 3)

    if Mouth[0] + Mouth[1] != 0:
        cv2.rectangle(sample, (Mouth[0], Mouth[1]), (Mouth[2], Mouth[3]), (255, 0, 0), 3)

    if Cigar[0] + Cigar[1] != 0:
        cv2.rectangle(sample, (Cigar[0], Cigar[1]), (Cigar[2], Cigar[3]), (255, 0, 255), 3)

    if Phone[0] + Phone[1] != 0:
        cv2.rectangle(sample, (Phone[0], Phone[1]), (Phone[2], Phone[3]), (255, 255, 0), 3)

    # 파일 저장
    cv2.imshow('sample', sample)
    while True:
        if cv2.waitKey() == 13: # ENTER 키
            break
        
        if cv2.waitKey() == 27: # ESC 키
            cv2.destroyAllWindows()
```
